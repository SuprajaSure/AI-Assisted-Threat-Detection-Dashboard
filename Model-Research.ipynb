# 1. Introduction

Cybersecurity has become one of the most critical aspects of modern organizations as digital systems continue to grow rapidly. Every day, computer networks generate a large number of security logs containing information about user activities, login attempts, network connections, and data transfers. Monitoring these logs manually is both time-consuming and difficult, especially when suspicious activities are hidden among thousands of normal events.

Artificial Intelligence (AI) provides an efficient way to analyze these security logs by automatically identifying unusual patterns that may indicate cyber threats. Instead of relying only on predefined security rules, AI algorithms can learn the normal behavior of network activities and detect deviations that could represent malicious attacks.

This project, **AI-Assisted Threat Detection Dashboard**, demonstrates how machine learning can be used to detect suspicious network behavior. The detected anomalies are then presented through a dashboard, allowing users to monitor network security more effectively.

# 5. Selected AI Model

## Isolation Forest

The machine learning model selected for this project is **Isolation Forest**.

Isolation Forest is an unsupervised machine learning algorithm designed specifically for anomaly detection. Unlike classification algorithms, it does not require labeled training data to identify suspicious activities.

Instead of learning what attacks look like, Isolation Forest learns what normal behavior looks like and isolates observations that are significantly different.

This makes it particularly suitable for cybersecurity because many cyber attacks appear as unusual or unexpected network activities.

---

# 6. Why Isolation Forest?

Isolation Forest was selected because it perfectly matches the objective of this project.

The goal of this project is not to classify different types of cyber attacks but to detect unusual activities that may indicate a potential threat.

Isolation Forest offers several advantages:

- Specifically designed for anomaly detection
- Works well with small datasets
- Does not require large amounts of labeled data
- Fast training and prediction
- Simple to implement using Python
- Detects unknown or previously unseen threats
- Produces anomaly scores that can easily be visualized on a dashboard

For a demonstration project using a small dataset, Isolation Forest provides an excellent balance between simplicity and performance.

---

# 7. How Isolation Forest Works

Isolation Forest identifies anomalies by isolating data points through random partitioning.

The working process is as follows:

1. Randomly select one feature from the dataset.
2. Randomly choose a split value for that feature.
3. Continue dividing the dataset into smaller partitions.
4. Repeat the process to build multiple decision trees.
5. Calculate how many splits are required to isolate each data point.
6. Data points that require fewer splits are considered anomalies because they are significantly different from the majority of the data.

Unlike many machine learning algorithms, Isolation Forest focuses on isolating abnormal observations rather than modeling normal behavior directly.
