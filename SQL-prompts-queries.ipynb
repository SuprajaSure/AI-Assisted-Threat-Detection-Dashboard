```python
import sqlite3
import pandas as pd

# Sample Network Traffic Dataset
data = {
    "Duration": [45, 12, 30, 8, 15, 60, 25, 18, 50, 35],
    "Protocol": ["TCP", "UDP", "TCP", "TCP", "UDP", "ICMP", "TCP", "UDP", "ICMP", "TCP"],
    "SourceIP": [
        "192.168.1.10", "192.168.1.15", "192.168.1.18", "192.168.1.20",
        "192.168.1.25", "192.168.1.30", "192.168.1.35", "192.168.1.40",
        "192.168.1.45", "192.168.1.50"
    ],
    "DestinationIP": [
        "10.0.0.5", "10.0.0.8", "10.0.0.12", "10.0.0.15",
        "10.0.0.18", "10.0.0.20", "10.0.0.22", "10.0.0.25",
        "10.0.0.28", "10.0.0.30"
    ],
    "SourcePort": [443, 53, 22, 8080, 53, 0, 80, 53, 0, 21],
    "DestinationPort": [80, 53, 443, 80, 53, 0, 8080, 53, 0, 22],
    "PacketCount": [120, 300, 210, 420, 380, 150, 350, 325, 180, 498],
    "ByteCount": [5600, 14500, 9800, 18000, 16000, 7000, 15200, 13800, 8100, 19500],
    "Label": [
        "Normal", "Attack", "Normal", "Attack", "Attack",
        "Normal", "Normal", "Attack", "Normal", "Attack"
    ]
}

df = pd.DataFrame(data)

# Create SQLite database
conn = sqlite3.connect(":memory:")

# Store DataFrame as SQL table
df.to_sql("NetworkTraffic", conn, index=False, if_exists="replace")

print("Database Created Successfully!")
```

SQL Task 1 – Display All Records
Prompt

Write an SQL query to display all records from the network traffic dataset.

```python
query = "SELECT * FROM NetworkTraffic;"
result = pd.read_sql(query, conn)
print(result)
```

SQL Task 2 – Display Only Attack Records
Prompt

Write an SQL query to display only the records labeled as Attack.

```python
query = """
SELECT *
FROM NetworkTraffic
WHERE Label='Attack';
"""

result = pd.read_sql(query, conn)
print(result)
```

SQL Task 3 – Count Total Records
Prompt

Write an SQL query to count the total number of network traffic records.

```python
query = """
SELECT COUNT(*) AS Total_Records
FROM NetworkTraffic;
"""

result = pd.read_sql(query, conn)
print(result)
```

SQL Task 4 – Count Normal and Attack Traffic
Prompt

Write an SQL query to count the number of records in each traffic category.

```python
query = """
SELECT Label,
COUNT(*) AS Total
FROM NetworkTraffic
GROUP BY Label;
"""

result = pd.read_sql(query, conn)
print(result)
```

SQL Task 5 – Find Maximum Packet Count
Prompt

Write an SQL query to find the maximum packet count.

```python
query = """
SELECT MAX(PacketCount) AS Maximum_PacketCount
FROM NetworkTraffic;
"""

result = pd.read_sql(query, conn)
print(result)
```

SQL Task 6 – Calculate Average Byte Count
Prompt

Write an SQL query to calculate the average byte count.

```python
query = """
SELECT AVG(ByteCount) AS Average_ByteCount
FROM NetworkTraffic;
"""

result = pd.read_sql(query, conn)
print(result)
```
